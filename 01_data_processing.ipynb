```python
def list_files(base_path, noExt=False):
    from os import listdir
    from os.path import isfile, join
    
    if noExt:
        onlyfiles = [f[0:-4] for f in listdir(base_path) if isfile(join(base_path, f))]
        return onlyfiles
    else:
        onlyfiles = [f for f in listdir(base_path) if isfile(join(base_path, f))]
        return onlyfiles

list_files(BASE_PATH)[:5]
```

```python
list_files(BASE_PATH, True)[:5]
```

```python
def load_covid_file(base_path, file_name_no_ext):
    df = pd.read_csv(base_path + file_name_no_ext + '.csv')
    country_col = 'Country_Region'
    if 'Country_Region' not in df.columns:
        country_col = 'Country/Region'
    
    df = df[[country_col, 'Confirmed', 'Deaths']]
    df.columns = ['Country_Region', 'Confirmed', 'Deaths']
    df = df.groupby(['Country_Region']).agg({
        'Confirmed':'sum', 
        'Deaths':'sum'
    }).reset_index()
    df['date'] = datetime.datetime.strptime(file_name_no_ext, '%m-%d-%Y').strftime('%m/%d/%Y')
    return df

df_test02 = load_covid_file(BASE_PATH, '01-01-2021')
df_test02[df_test02['Country_Region'] == 'United Kingdom']
```

### 4.4 Validation

```python
dfs_merged_regions = pd.concat(list(dfs_by_region.values())).reset_index(drop=True)
dfs_merged_regions
```

```python
list_unique_codes = dfs_merged_regions['Short Code'].unique()
list_unique_codes
```

```python
# sanity check
dfs_merged_regions[dfs_merged_regions['Short Code'] == 'CHN'].sort_values(by=['date'])
```

```python
# validation

for i in range(len(list_unique_codes)):
    current_code = list_unique_codes[i]
    df_current_by_code = dfs_merged_regions[dfs_merged_regions['Short Code'] == current_code]
    
    print('checking: {} (total {})'.format(current_code, len(df_current_by_code)))
    
    current_unique_dates = df_current_by_code['date'].unique()
    
    for j in range(len(current_unique_dates)):
        current_date = current_unique_dates[j]
        df_current_by_code_and_date = df_current_by_code[df_current_by_code['date'] == current_date]
        
        if len(df_current_by_code_and_date) != 1:
            print('    error at date: ' + current_date)
```

```python
# remove regions with too small amount of data
dfs_merged_regions = dfs_merged_regions.drop(dfs_merged_regions[dfs_merged_regions['Short Code'] == 'CHI'].index)
dfs_merged_regions = dfs_merged_regions.drop(dfs_merged_regions[dfs_merged_regions['Short Code'] == 'FRO'].index)
dfs_merged_regions = dfs_merged_regions.drop(dfs_merged_regions[dfs_merged_regions['Short Code'] == 'GIB'].index)
dfs_merged_regions = dfs_merged_regions.drop(dfs_merged_regions[dfs_merged_regions['Short Code'] == 'ABW'].index)
dfs_merged_regions = dfs_merged_regions.drop(dfs_merged_regions[dfs_merged_regions['Short Code'] == 'GRL'].index)
dfs_merged_regions = dfs_merged_regions.drop(dfs_merged_regions[dfs_merged_regions['Short Code'] == 'GUM'].index)
dfs_merged_regions = dfs_merged_regions.drop(dfs_merged_regions[dfs_merged_regions['Short Code'] == 'PRI'].index)
dfs_merged_regions = dfs_merged_regions.drop(dfs_merged_regions[dfs_merged_regions['Short Code'] == 'CYM'].index)
dfs_merged_regions = dfs_merged_regions.drop(dfs_merged_regions[dfs_merged_regions['Short Code'] == 'CUW'].index)
dfs_merged_regions = dfs_merged_regions.drop(dfs_merged_regions[dfs_merged_regions['Short Code'] == 'PRK'].index)

dfs_merged_regions = dfs_merged_regions.reset_index(drop=True)
```

```python
dfs_merged_regions
```

```python
# run validation again
list_unique_codes = dfs_merged_regions['Short Code'].unique()

for i in range(len(list_unique_codes)):
    current_code = list_unique_codes[i]
    df_current_by_code = dfs_merged_regions[dfs_merged_regions['Short Code'] == current_code]
    
    print('checking: {} (total {})'.format(current_code, len(df_current_by_code)))
    
    current_unique_dates = df_current_by_code['date'].unique()
    
    for j in range(len(current_unique_dates)):
        current_date = current_unique_dates[j]
        df_current_by_code_and_date = df_current_by_code[df_current_by_code['date'] == current_date]
        
#         print(df_current_by_code_and_date)
        
        if len(df_current_by_code_and_date) != 1:
            print('    error at date: ' + current_date)
```
